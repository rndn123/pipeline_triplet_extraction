```python
def minimum(tag1, tag2):
    tags = []
    for i, x in enumerate(tag1):
        tags.append(min(x, tag2[i]))
    return tags
```

```python
count = 0
for i in range(len(df)):
    task = df.iloc[i,2]
    pid = df.iloc[i,3]
    sent_id = df.iloc[i, 4]
    start_id, end_id, phrases = [], [], []
    words = load(df.iloc[i, 5])
    tags = load(df.iloc[i, 8])
    #tag2 = load(df.iloc[i, 9])
    #tags = minimum(tag1, tag2)
    start, end = 0, 0
    phrase = ''
    for i, x in enumerate(tags):
        if(i >= len(words)):
            break
        if(x == 2 and start == end):
            end += len(words[i]) + 1
            start = end
            continue
        if(x == 2):
            start_id.append(start)
            end_id.append(end-1)
            phrases.append(phrase)
            end += len(words[i]) + 1
            start = end
            phrase = ''
        if(x == 1):
            if(end > start):
                phrase += " "
            phrase += words[i]
            end += len(words[i]) + 1
        if(x == 0 and start == end):
            phrase += words[i]
            end += len(words[i]) + 1
            continue
        if(x == 0):
            start_id.append(start)
            end_id.append(end-1)
            phrases.append(phrase)
            start = end
            end += len(words[i]) + 1
            phrase = words[i]
    if(start != end):
        start_id.append(start)
        end_id.append(end-1)
        phrases.append(phrase)
    path = task + "/" + str(pid) + "/entities.txt"
    with open(path, mode='at', encoding='utf-8') as file:
        for i, x in enumerate(phrases):
            file.write("%i\t%i\t%i\t%s\n" %(sent_id, start_id[i], end_id[i], x))
    path = task + "/" + str(pid) + "/sentences.txt"
    with open(path, mode='at', encoding='utf-8') as file:
        file.write("%i\n" % sent_id)
print(count)
```
