```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/PubMed/train_pos_pubmed_238.json")
df2 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/PubMed/val_pos_pubmed_23.json")
df3 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/PubMed/train_neg_pubmed_46.json")
df4 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/PubMed/val_neg_pubmed_4.json")
df5 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/PubMed/test_pubmed_100.json")
df6 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/additonal/manual_perturbations_19.json")
df7 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/CORD-19/train_pos_cord_117.json")
df8 = pd.read_json("https://raw.githubusercontent.com/siftech/SciClaim/main/CORD-19/val_pos_cord_18.json")
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8])
```

```python
df.drop(columns = ['uuid', 'notes', 'modified', 'dataset', 'relations', 'attributes', 'user'], inplace = True)
```

```python
df
```

```python
import nltk
def tag(tokens):
    tags = nltk.pos_tag(tokens)
    tags = [x[1] for x in tags]
    return tags
```

```python
from ast import literal_eval as load
```

```python
sentence_id = []
words = []
labels = []
pos_tag = []
for i in range(len(df)):
    tokens = df.iloc[i, 1]
    entities = df.iloc[i, 2]
    label = ['O']*len(tokens)
    sid = [i]*len(tokens)
    for entity in entities:
        x = entity['start']
        label[x] = 'B'
        x = x+1
        while(x<entity['end']):
            label[x] = 'I'
            x = x+1
    tags = tag(tokens)
    sentence_id.extend(sid)
    words.extend(tokens)
    labels.extend(label)
    pos_tag.extend(tags)
```

```python
df_covid = pd.DataFrame({'sentence_id': sentence_id, 'words': words, 'labels': labels, 'pos_tag':pos_tag})
```

```python
df_covid.to_csv("Dataset_TaskB/SciClaim.csv", index = None)
```
