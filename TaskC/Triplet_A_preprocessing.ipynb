```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset_TaskC/Test_Results_Triplets2_A.csv")
```

```python
df
```

```python
input_sent = []
for i in range(len(df)):
    triplets = []
    for j in range(3):
        triplets.append([df.iloc[i, 5+5*j], df.iloc[i, 6+5*j], df.iloc[i, 8+5*j]])
    triplets = sorted(triplets)
    sent = df.iloc[i, 4]
    for j, triplet in enumerate(triplets):
        if(triplet[2] == 0):
            ts, te = "[[ ", " ]]"
        else:
            ts, te = "<< ", " >>"
        x = triplet[0] + 6*j
        sent = sent[:x] + ts + sent[x:]
        x = triplet[1] + 6*j + 3
        while(x<len(sent) and sent[x]!=' '):
            x += 1
        sent = sent[:x] + te + sent[x:]
    input_sent.append(sent)
df['input'] = input_sent
```
