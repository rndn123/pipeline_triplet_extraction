```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset_TaskC/Test_Results_Triplets2_B.csv")
```

```python
input_sent = []
for i in range(len(df)):
    triplets = []
    triplets.append([df.iloc[i, 5], df.iloc[i, 6]])
    triplets.append([df.iloc[i, 10], df.iloc[i, 11]])
    sent = df.iloc[i, 4]
    for j, triplet in enumerate(triplets):
        if(j == 1):
            ts, te = "[[ ", " ]]"
        else:
            ts, te = "<< ", " >>"
        x = triplet[0] + 6*j
        sent = sent[:x] + ts + sent[x:]
        x = triplet[1] + 6*j + 3
        while(x<len(sent) and sent[x]!=' '):
            x += 1
        sent = sent[:x] + te + sent[x:]
    input_sent.append(sent)
df['input'] = input_sent
```

```python
df = pd.merge(df, dfg[dfg['pred'] == 'has'], on = ['topic', 'paper_ID', 'info-unit', 'sub', 'obj'], how = 'left', indicator=True)
```

```python
df['labels'] = df['_merge'].replace(label_dict)
```
