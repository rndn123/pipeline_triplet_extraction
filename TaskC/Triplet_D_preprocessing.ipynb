```python
input_sent = []
subject = []
for i in range(len(df)):
    x = df.iloc[i, 5]
    y = df.iloc[i, 6] + 3
    sent = df.iloc[i, 4]
    info = df.iloc[i, 3]
    info = info.replace('-', ' ')
    x = df.iloc[i, 5]
    sent = sent[:x] + "[[ " + sent[x:]
    x = df.iloc[i, 6] + 3
    while(x<len(sent) and sent[x]!=' '):
        x += 1
    sent = sent[:x] + " ]]" + sent[x:]
    sent = "[[ " + info + " ]] : " + sent
    input_sent.append(sent)
    subject.append(info)
df['sub'] = subject
df['input'] = input_sent
```
