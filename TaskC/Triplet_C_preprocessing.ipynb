```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset_TaskC/Test_Results_Triplets2_C.csv")
```

```python
df
```

```python
input_sent = []
subject = []
for i in range(len(df)):
    triplets = []
    for j in range(2):
        triplets.append([df.iloc[i, 5+5*j], df.iloc[i, 6+5*j], df.iloc[i, 8+5*j]])
    triplets = sorted(triplets)
    sent = df.iloc[i, 4]
    info = df.iloc[i, 3]
    info = info.replace('-', ' ')
    for j, triplet in enumerate(triplets):
        if(triplet[2] == 0):
            ts, te = "[[ ", " ]]"
        else:
            ts, te = "<< ", " >>"
        x = triplet[0] + 6*j
        sent = sent[:x] + ts + sent[x:]
        x = triplet[1] + 6*j + 3
        while(x<len(sent) and sent[x]!=' '):
            x += 1
        sent = sent[:x] + te + sent[x:]
    sent = "[[ " + info + " ]] : " + sent
    input_sent.append(sent)
    subject.append(info)
df['sub'] = subject
df['input'] = input_sent
```

```python
df
```

```python
dfg = pd.read_csv("Dataset_TaskC/Test_Ground_Triplets.csv")
dfg = dfg[['topic', 'paper_ID', 'info-unit', 'sub', 'pred', 'obj']]
```

```python
dfg['sub'] = dfg['sub'].str.lower()
dfg
```

```python
df = pd.merge(df, dfg, on = ['topic', 'paper_ID', 'info-unit', 'sub', 'pred', 'obj'], how = 'left', indicator=True)
```
