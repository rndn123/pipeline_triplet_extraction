```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset_TaskC/Test_Results_TaskC2.csv")
```

```python
num = []
pre, paper, sent_id, n = "", 0, 0, 0
for i in range(len(df)):
    cur = df.iloc[i, 0]
    pid = df.iloc[i, 1]
    sid = df.iloc[i, 2]
    if (cur == pre and pid == paper and sid == sent_id):
        n += 1
    else:
        n = 0
    num.append(n)
    pre, paper, sent_id = cur, pid, sid
df['num'] = num
```

```python
df.rename(columns = {'start_index': 'start_sub', 'end_index': 'end_sub', 'phrases': 'sub', 'labels': 'labels_sub', 'num': 'sub_num'}, inplace = True)
```

```python
df_pred = df.rename(columns = {'start_sub': 'start_pred', 'end_sub': 'end_pred', 'sub': 'pred', 'labels_sub': 'labels_pred', 'sub_num': 'pred_num'})
```
